<h1 id="item1">I. Download and Explore Dataset<h1>

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

<ol start="2"><h3><li>Extract the data from Website</li></h3></ol>

We will start by viewing our HTML code

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "lxml")
# We print the data to see the html code
soup.body.table
```

Then we extract the headings of the table

```python
table = soup.find("table", attrs={"class": "wikitable sortable"})
table_data = table.tbody.find_all("tr") 
heading = []
for th in table_data[0].find_all("th"):
    # remove any newlines and extra spaces from left and right
    heading.append(th.text.replace('\n', ' ').strip())

heading # print the heading
```

Then we extract the main data of the table and instert it into a dictionary

```python
n=len(heading)
data={}
data[heading[0]]=[]
data[heading[1]]=[]
data[heading[2]]=[]
for tr in table_data :
    for td,i in zip(tr.find_all("td"),range(n)) :
        data[heading[i]].append(td.text.replace('\n', ' ').strip())
data # data is dictionary that contains the data on every column e.g {'Postal code':['M1A',...],'Borough':['North York'],...}
```

<ol start="3"><h3><li>Now we transform the data into Pandas dataframe and clean our data</li></h3></ol>

<ul><li><h4>Columns name</h4></li></ul>

```python
df_toronto=pd.DataFrame(data)
# We change the column name Postal code to PostalCode
df_toronto.rename(columns={'Postal code':'PostalCode'},inplace=True)
df_toronto.head()
```

<ul><li><h4>We will Ignore cells with a borough that is Not assigned</h4></li></ul>

```python
df_toronto=df_toronto[df_toronto['Borough']!='Not assigned']
df_toronto.reset_index(inplace=True,drop=True)
df_toronto.head()
```

<ul><li><h4>Concatenate rows with the same PostalCode value</h4></li></ul>

```python
df_toronto= df_toronto.groupby('PostalCode',sort=False).agg({'Borough':'first', 
                             'Neighborhood': ', '.join}).reset_index()
# We replace existing / to ,
df_toronto['Neighborhood'].replace(' /',',',regex=True,inplace=True)
df_toronto.head()
```

<ul><li><h4>Replace 'Not assigned' values in Neighborhood with the Borough name</h4></li></ul>

```python

df_toronto.Neighborhood.replace('Not assigned',df_toronto.Borough,inplace=True)
df_toronto.head()
```

<ul><li><h4>Dataframe shape</h4></li></ul>

```python
df_toronto.shape
```

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

#### Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_grouped.shape
```

Now let's create the new dataframe and display the top 10 venues for each neighborhood.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
toronto_venues_sorted = pd.DataFrame(columns=columns)
toronto_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    toronto_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

toronto_venues_sorted.head()
```

Before start runing the *k*-means algorithm we should chose the number k

```python
df_toronto['Borough'].value_counts()
```

*k*= 4 is the optimum for our clustering algorithm

Run *k*-means to cluster the neighborhood into 4 clusters.

```python
from sklearn.cluster import KMeans
# set number of clusters
kclusters = 4

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped.drop('Neighborhood', axis=1))

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
# add clustering labels
toronto_venues_sorted.insert(0, 'Cluster', kmeans.labels_)


# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
df_toronto = df_toronto.join(toronto_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

df_toronto.head() # check the last columns!
```

Let's visualize the cluster map

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighborhood'], df_toronto['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
